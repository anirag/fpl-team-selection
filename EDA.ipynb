```python
import requests, pickle, shutil, time

all_players = {}
for i in range(600):
    playerurl = f"http://fantasy.premierleague.com/web/api/elements/{i}/"
    r = requests.get(playerurl)

    # skip non-existent players
    if r.status_code != 200:
        continue

    try:
        all_players[i] = r.json()
    except ValueError:
        continue
```

```python
outfile = open("all_players_data.pickle", "wb")
pickle.dump(all_players, outfile)
```
